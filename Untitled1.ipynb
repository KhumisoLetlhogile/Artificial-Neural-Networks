# Relevant Functions

```python
def init_params():
    W_i_h = np.random.uniform(-0.5,0.5,(10,x_train.shape[0]))
    b_i_h = np.zeros((10,1))
    W_h_o = np.random.uniform(-0.5,0.5,(2,10))
    b_h_o = np.zeros((2,1))
    return  W_i_h, b_i_h, W_h_o, b_h_o
```

### Sigmoid Function  
### $f(x) = \frac{1}{1+e^x}$

### Softmax Function
### $f(x_i)= \frac{e^{x_i}}{\sum_{j=1}^{K}e^{x_j}}$


```python
def sigmoid(Z):
    return 1/(1 + np.exp(-Z))

def softmax(Z):
    return np.exp(Z)/np.sum(np.exp(Z))
  
def ReLu(Z):
    return np.maximum(0,Z)

def deriv_ReLu(Z):
    return Z >0
```

```python
def Forward_Propagate(W_i_h, b_i_h , W_h_o, b_h_o,X):
    H_pred = W_i_h.dot(X) + b_i_h
    H = ReLu(H_pred)
    O_pred =  W_h_o.dot(H) + b_h_o
    O = softmax(O_pred)
    return H_pred, H, O_pred, O
```

```python
def one_hot(Y):
    one_hot_Y = np.zeros((Y.size,int(Y.max()+1)))
    one_hot_Y[np.arange(Y.size),Y.astype(int)] = 1
    one_hot_Y = one_hot_Y.T
    return one_hot_Y
                          

def Backward_Propagate(H_pred,H,O_pred,O,W_h_o,X,Y):
    m = Y.size
    one_hot_Y = one_hot(Y)
    dO_pred = O - one_hot_Y
    dW_h_o = 1/m*dO_pred.dot(H.T)
    dB_h_o = 1/m*np.sum(dO_pred)
    dH_pred = W_h_o.T.dot(dO_pred)*deriv_ReLu(H_pred)
    dW_i_h = 1/m*dH_pred.dot(X.T)
    dB_i_h = 1/m*np.sum(dH_pred)
    return dW_i_h, dB_i_h,dW_h_o,dB_h_o

def update_params( W_i_h, b_i_h, W_h_o, b_h_o,dW_i_h, dB_i_h,dW_h_o,dB_h_o, alpha):
    W_i_h = W_i_h - alpha*dW_i_h
    b_i_h = b_i_h - alpha*dB_i_h
    W_h_o = W_h_o - alpha*dW_h_o
    b_h_o = b_h_o - alpha*dB_h_o 
    return W_i_h, b_i_h, W_h_o, b_h_o
```

```python
def get_predictions(O):
    return np.argmax(O,0)
def get_accuracy(predictions,Y):
    print(predictions,Y)
    return np.sum(predictions == Y)/Y.size
```

```python
def gradient_descent(X,Y,iterations, alpha):
    W_i_h, b_i_h, W_h_o, b_h_o = init_params()
    for i in range(iterations):
        H_pred, H, O_pred, O = Forward_Propagate(W_i_h, b_i_h , W_h_o, b_h_o,X)
        dW_i_h, dB_i_h,dW_h_o,dB_h_o = Backward_Propagate(H_pred,H,O_pred,O,W_h_o,X,Y)
        W_i_h, b_i_h, W_h_o, b_h_o = update_params( W_i_h, b_i_h, W_h_o, b_h_o,dW_i_h, dB_i_h,dW_h_o,dB_h_o, alpha)
        if(i% 2==0):
            print("Iteration: ", i)
            print("Accuracy: ", get_accuracy(get_predictions(O),Y))
    return W_i_h, b_i_h, W_h_o, b_h_o
```
